```python
d1={
    "a":[1,2,3,4,5,6],
    "b":[7,8,9,10,11,12],
    
}
d2={
    "a":[1,2,3,7,8,9],
    "c":[7,8,9,10,11,12],
    
}
df1=pd.DataFrame(d1,index=[1,2,3,4,5,6])
df2=pd.DataFrame(d2,index=[1,2,3,7,8,9])
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
# merge works based on column default return inner join
# df1 and df2 has same index ,so based on common col name it returns common values ,it returns inner join values
```
