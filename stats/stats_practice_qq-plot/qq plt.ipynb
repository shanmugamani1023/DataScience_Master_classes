```python
# Q-Q plot used to check particular distribution is whether normal or not
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pylab
import scipy.stats as stat
```

```python
data=sb.load_dataset('iris')
data
```

```python
def check(data,col_name):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sb.histplot(data[col_name],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(data[col_name],dist='norm',plot=pylab)
```

```python
check(data,'sepal_length')
```
