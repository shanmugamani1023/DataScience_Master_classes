# Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook

```python
from flask import Flask,jsonify
app=Flask(__name__)

@app.route("/")
def hello():
    return "hello world"


@app.route("/data")
def data():
    result={
        "name":"shan",
        "job":"data scientist"
        
    }
    return jsonify(result)

if __name__=="__main__":
    app.run()
```
