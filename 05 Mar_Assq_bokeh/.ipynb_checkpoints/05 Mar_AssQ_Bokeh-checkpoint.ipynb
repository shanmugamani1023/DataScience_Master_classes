# Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, output_file, show

# define some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# create a Bokeh figure with the desired properties
p = figure(title="My Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis', plot_width=400, plot_height=400)

# add circles to the figure using the data and glyph properties
p.circle(x, y, size=10, color='blue')

# display the plot in a web browser or save it to an HTML file
output_file('my_plot.html')
show(p)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example

# In Bokeh, glyphs are visual shapes used to represent data points in a plot. Glyphs can be added to a Bokeh plot using the glyph functions provided by the bokeh.plotting module, such as circle, line, rect, and image. Glyph functions take as input the data to be plotted and various glyph properties, such as size, color, and line width, and return a glyph object that can be added to a Bokeh figure.

Here's an example of adding glyphs to a Bokeh plot:

```python
from bokeh.plotting import figure, output_file, show

# define some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ['red', 'green', 'blue', 'orange', 'purple']
sizes = [10, 20, 30, 40, 50]

# create a Bokeh figure with the desired properties
p = figure(title="My Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis', plot_width=400, plot_height=400)

# add circles to the figure using the data and glyph properties
p.circle(x, y, size=sizes, fill_color=colors, line_color='black', line_width=2)

# add a line to the figure
p.line(x, y, line_color='gray', line_width=2)

# display the plot in a web browser or save it to an HTML file
output_file('my_plot.html')
show(p)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure, output_file, show

p = figure(title="My Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
```

```python
p.circle(x, y, size=10, color='blue', legend_label='Data Points')
p.legend.location = 'top_left'
```

```python
p.circle(x, y, size=10, color='blue', line_color='black', line_width=2)
p.line(x, y, line_color='gray', line_width=2, line_dash=[4, 4])
```

# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

# Bokeh server is a way to create interactive Bokeh plots that can be updated in real-time, directly in a web browser. It allows users to interact with Bokeh plots in a web application, and enables the creation of dynamic, data-driven dashboards and applications.

To use the Bokeh server, you need to write a Python script that defines the layout and behavior of your application, and then start the Bokeh server with that script. The script will typically contain a curdoc() function, which represents the current document being displayed by the server. You can add widgets, plots, and other interactive elements to the document, and define callbacks that respond to user interactions.
