# Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
# pdf= EXPONENET_value*CO-EFFICIENT
import math
mu=1
sigma=2
x=1
```

```python
exponent=-((x-mu)**2/sigma**2)
exponent_value=math.exp(exponent)

coeffcient=1/(sigma*math.sqrt(2*math.pi))
pdf=exponent_value*coeffcient
print(pdf)
```

```python
from scipy.stats import norm

norm.pdf(x,mu,sigma)
```

# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
sample=np.random.binomial(n=1,p=0.4,size=1000)
sb.histplot(sample,kde=True)
plt.title('Binomial Distribution with p=0.4')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(1234)

# Generate a random sample of size 1000 from a binomial distribution with p=0.4
sample = np.random.binomial(n=1, p=0.4, size=1000)

# Plot a histogram of the sample using matplotlib
plt.hist(sample, edgecolor='black')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Binomial Distribution with p=0.4')
plt.show()
```

# Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
np.random.seed(1)
sample=np.random.poisson(lam=5,size=1000)
mean=sample.mean()
var=sample.var()
print(mean,var)
plt.hist(sample)
plt.xlabel('poisson values')
plt.ylabel('count')
plt.title('poisson')
plt.show()
```

# Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

# In a normal distribution, the least frequent data appears at the tails of the distribution, which are the regions furthest away from the mean. Specifically, the probability density function of a normal distribution is highest at the mean and decreases symmetrically in both directions away from the mean. This means that the values far away from the mean have a lower probability of occurring than the values closer to the mean. 

In other words, the tails of the distribution contain the values that are farthest from the mean, and these values are less likely to occur than values closer to the mean. Therefore, the least frequent data in a normal distribution appear in the tails, which are the regions farthest away from the mean.
