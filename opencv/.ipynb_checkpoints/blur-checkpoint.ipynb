```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# # blur operation
# 1.gaussian fillter
# 2.median 
# 3.bilatteral fillter
```

```python
image=cv2.imread("images/noisy.png")
```

```python
blur=cv2.blur(image,(5,5))
```

```python
gaussian=cv2.GaussianBlur(image,(5,5),10)
gaussian_0=cv2.GaussianBlur(image,(5,5),0)
```

```python
median=cv2.medianBlur(image,5)
# gaussian_0=cv2.GaussianBlur(image,(5,5),0)
```

```python
cv2.imshow("input",image)
cv2.imshow("blur",blur)
cv2.imshow("gaussian",gaussian)
cv2.imshow("gaussian_0",gaussian_0)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
