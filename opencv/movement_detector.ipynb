```python
import numpy as np
import cv2
```

```python
cap=cv2.VideoCapture(1)
ret2,frame2=cap.read()


while True:
    ret1,frame1=cap.read()
    gray1=cv2.cvtColor(frame1,cv2.COLOR_BGR2GRAY)
    
    blur1=cv2.GaussianBlur(gray1,(11,11),0)
    
    thresh1=cv2.threshold(blur1,80,255,cv2.THRESH_BINARY_INV)
    white_pixels=cv2.countNonZero(thresh1[1])
    print(white_pixels)
    
    width=frame1.shape[1]
    height=frame1.shape[0]
    total=width*height
    
    if white_pixels > 60047:
        print("wow")
    
    
    
      
    cv2.imshow("frame1",thresh1[1])
    key=cv2.waitKey(10)
    if key==ord("q"):
        break
cv2.destroyAllWindows()

    
    
```
