# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset

```python
import seaborn as sns
```

```python
data=sns.load_dataset("fmri")
```

```python
sns.lineplot(x=data.timepoint,y=data.signal)
```

```python
sns.violinplot(x=data.timepoint,y=data.signal)
```

```python
sns.pairplot(data)
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.   Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset

```python
data_titanic=sns.load_dataset("titanic")
```

```python
sns.boxplot(x=data_titanic.pclass,y=data_titanic.age)
```

```python
sns.boxplot(x=data_titanic.pclass,y=data_titanic.fare)
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset. 

```python
d2=sns.load_dataset("diamonds")
```

```python
sns.histplot(d2["price"],)
```

```python
sns.relplot(x=d2.price,y=d2.cut,hue='cut',data=d2)
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
d3=sns.load_dataset("iris")
sns.pairplot(d3,hue="species")
```
